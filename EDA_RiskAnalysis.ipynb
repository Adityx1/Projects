```python

# Loan Default Risk Analysis - EDA Implementation

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from scipy import stats
from sklearn.preprocessing import LabelEncoder
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Configuration
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

# SECTION 1: DATA LOADING AND INITIAL EXPLORATION

def load_and_explore_data():
    """Load datasets and perform initial exploration"""
    print("=" * 60)
    print("LOAN DEFAULT RISK ANALYSIS - EDA")
    print("=" * 60)

    # Load datasets
    try:
        app_data = pd.read_csv('application_data.csv')
        prev_data = pd.read_csv('previous_application.csv')
        print(" Data loaded successfully")
    except FileNotFoundError as e:
        print(f" Error loading data: {e}")
        return None, None

    # Basic information
    print(f"\n DATASET OVERVIEW:")
    print(f"Application data shape: {app_data.shape}")
    print(f"Previous data shape: {prev_data.shape}")

    print(f"\n TARGET VARIABLE DISTRIBUTION:")
    target_dist = app_data['TARGET'].value_counts()
    target_pct = app_data['TARGET'].value_counts(normalize=True) * 100

    print(f"Class 0 (No Payment Difficulties): {target_dist[0]:,} ({target_pct[0]:.2f}%)")
    print(f"Class 1 (Payment Difficulties): {target_dist[1]:,} ({target_pct[1]:.2f}%)")
    print(f"Imbalance Ratio: {target_dist[0]/target_dist[1]:.1f}:1")

    return app_data, prev_data


# SECTION 2: MISSING DATA ANALYSIS


def analyze_missing_data(df):
    """Comprehensive missing data analysis"""

    print("MISSING DATA ANALYSIS")


    # Calculate missing percentages
    missing_percent = (df.isnull().sum() / len(df)) * 100
    missing_df = pd.DataFrame({
        'Column': missing_percent.index,
        'Missing_Count': df.isnull().sum(),
        'Missing_Percentage': missing_percent.values,
        'Data_Type': df.dtypes
    }).sort_values('Missing_Percentage', ascending=False)

    # Filter columns with missing data
    missing_df = missing_df[missing_df['Missing_Percentage'] > 0]

    # Categorize by missing percentage
    high_missing = missing_df[missing_df['Missing_Percentage'] > 50]
    medium_missing = missing_df[(missing_df['Missing_Percentage'] >= 10) &
                               (missing_df['Missing_Percentage'] <= 50)]
    low_missing = missing_df[(missing_df['Missing_Percentage'] > 0) &
                            (missing_df['Missing_Percentage'] < 10)]

    print(f" MISSING DATA SUMMARY:")
    print(f"Total columns: {len(df.columns)}")
    print(f"Columns with missing data: {len(missing_df)}")
    print(f"High missing (>50%): {len(high_missing)}")
    print(f"Medium missing (10-50%): {len(medium_missing)}")
    print(f"Low missing (<10%): {len(low_missing)}")

    # Visualization
    plt.figure(figsize=(15, 8))
    top_missing = missing_df.head(20)

    plt.subplot(2, 2, 1)
    plt.barh(range(len(top_missing)), top_missing['Missing_Percentage'])
    plt.yticks(range(len(top_missing)), top_missing['Column'])
    plt.xlabel('Missing Percentage')
    plt.title('Top 20 Columns with Missing Data')
    plt.gca().invert_yaxis()

    plt.subplot(2, 2, 2)
    categories = ['High (>50%)', 'Medium (10-50%)', 'Low (<10%)']
    counts = [len(high_missing), len(medium_missing), len(low_missing)]
    plt.pie(counts, labels=categories, autopct='%1.1f%%')
    plt.title('Missing Data Categories')

    plt.tight_layout()
    plt.show()

    return missing_df, high_missing, medium_missing, low_missing

def handle_missing_data(df, high_missing_cols, strategy='auto'):
    """Handle missing data based on strategy"""
    df_clean = df.copy()

    print(f"\n MISSING DATA TREATMENT:")

    # Drop high missing columns (>50%)
    if len(high_missing_cols) > 0:
        cols_to_drop = high_missing_cols['Column'].tolist()
        df_clean = df_clean.drop(columns=cols_to_drop)
        print(f"Dropped {len(cols_to_drop)} columns with >50% missing data")

    # Handle remaining missing data
    for col in df_clean.columns:
        if df_clean[col].isnull().sum() > 0:
            if df_clean[col].dtype == 'object':
                # Categorical: fill with mode
                mode_val = df_clean[col].mode()[0] if len(df_clean[col].mode()) > 0 else 'Unknown'
                df_clean[col].fillna(mode_val, inplace=True)
            else:
                # Numerical: fill with median
                median_val = df_clean[col].median()
                df_clean[col].fillna(median_val, inplace=True)

    print(f" Missing data handled. New shape: {df_clean.shape}")
    return df_clean


# SECTION 3: OUTLIER DETECTION AND ANALYSIS

def detect_outliers(df, method='iqr'):
    """Detect outliers in numerical columns"""

    print("OUTLIER DETECTION")


    numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    if 'TARGET' in numerical_cols:
        numerical_cols.remove('TARGET')

    outlier_summary = []

    for col in numerical_cols[:10]:  # Analyze top 10 numerical columns
        if method == 'iqr':
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
            outlier_count = len(outliers)

        elif method == 'zscore':
            z_scores = np.abs(stats.zscore(df[col].dropna()))
            outliers = df[z_scores > 3]
            outlier_count = len(outliers)

        outlier_percentage = (outlier_count / len(df)) * 100

        outlier_summary.append({
            'Column': col,
            'Outlier_Count': outlier_count,
            'Outlier_Percentage': outlier_percentage,
            'Mean': df[col].mean(),
            'Median': df[col].median(),
            'Std': df[col].std()
        })

    outlier_df = pd.DataFrame(outlier_summary)
    outlier_df = outlier_df.sort_values('Outlier_Percentage', ascending=False)

    print(f" OUTLIER SUMMARY (Top 10 Numerical Columns):")
    print(outlier_df.to_string(index=False))

    # Visualize outliers for top 5 columns
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    axes = axes.ravel()

    for i, col in enumerate(outlier_df.head(5)['Column']):
        # Box plot
        axes[i].boxplot(df[col].dropna())
        axes[i].set_title(f'{col}\nOutliers: {outlier_df[outlier_df["Column"]==col]["Outlier_Percentage"].iloc[0]:.1f}%')
        axes[i].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()

    return outlier_df


# SECTION 4: UNIVARIATE ANALYSIS

def univariate_analysis(df, target_col='TARGET'):
    """Comprehensive univariate analysis"""

    print("UNIVARIATE ANALYSIS")


    # Separate numerical and categorical columns
    numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

    if target_col in numerical_cols:
        numerical_cols.remove(target_col)

    print(f" Analyzing {len(numerical_cols)} numerical and {len(categorical_cols)} categorical variables")

    # Numerical variables analysis
    print(f"\n NUMERICAL VARIABLES SUMMARY:")
    numerical_summary = df[numerical_cols[:10]].describe()
    print(numerical_summary)

    # Plot distributions for top numerical variables
    fig, axes = plt.subplots(3, 3, figsize=(18, 15))
    axes = axes.ravel()

    for i, col in enumerate(numerical_cols[:9]):
        axes[i].hist(df[col].dropna(), bins=50, alpha=0.7, edgecolor='black')
        axes[i].set_title(f'Distribution of {col}')
        axes[i].set_xlabel(col)
        axes[i].set_ylabel('Frequency')

    plt.tight_layout()
    plt.show()

    # Categorical variables analysis
    print(f"\n CATEGORICAL VARIABLES ANALYSIS:")

    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    axes = axes.ravel()

    for i, col in enumerate(categorical_cols[:4]):
        if df[col].nunique() <= 20:  # Only plot if reasonable number of categories
            value_counts = df[col].value_counts()
            axes[i].bar(range(len(value_counts)), value_counts.values)
            axes[i].set_title(f'Distribution of {col}')
            axes[i].set_xlabel(col)
            axes[i].set_ylabel('Count')
            axes[i].tick_params(axis='x', rotation=45)

            # Add value labels
            for j, v in enumerate(value_counts.values):
                axes[i].text(j, v + max(value_counts.values)*0.01, str(v), ha='center')

    plt.tight_layout()
    plt.show()


# SECTION 5: BIVARIATE ANALYSIS


def bivariate_analysis(df, target_col='TARGET'):
    """Bivariate analysis with target variable"""

    print("BIVARIATE ANALYSIS")


    numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

    if target_col in numerical_cols:
        numerical_cols.remove(target_col)

    # Numerical vs Target
    print(f" NUMERICAL VARIABLES VS TARGET:")

    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    axes = axes.ravel()

    for i, col in enumerate(numerical_cols[:6]):
        # Box plot by target
        df.boxplot(column=col, by=target_col, ax=axes[i])
        axes[i].set_title(f'{col} by Payment Difficulties')
        axes[i].set_xlabel('Payment Difficulties (0=No, 1=Yes)')
        axes[i].set_ylabel(col)

    plt.tight_layout()
    plt.show()

    # Statistical comparison
    print(f"\n STATISTICAL COMPARISON BY TARGET:")
    for col in numerical_cols[:5]:
        print(f"\n{col}:")
        print(df.groupby(target_col)[col].agg(['mean', 'median', 'std']).round(2))

    # Categorical vs Target
    print(f"\n CATEGORICAL VARIABLES VS TARGET:")

    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    axes = axes.ravel()

    for i, col in enumerate(categorical_cols[:4]):
        if df[col].nunique() <= 10:  # Only for manageable categories
            # Cross-tabulation with percentages
            ct = pd.crosstab(df[col], df[target_col], normalize='index') * 100
            ct.plot(kind='bar', ax=axes[i])
            axes[i].set_title(f'{col} vs Payment Difficulties (%)')
            axes[i].set_xlabel(col)
            axes[i].set_ylabel('Percentage')
            axes[i].legend(['No Difficulties', 'Payment Difficulties'])
            axes[i].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()

# SECTION 6: CORRELATION ANALYSIS


def correlation_analysis(df, target_col='TARGET', top_n=10):
    """Find top correlations for each target class"""

    print("CORRELATION ANALYSIS")


    # Separate data by target
    target_0 = df[df[target_col] == 0]
    target_1 = df[df[target_col] == 1]

    # Get numerical columns only
    numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    if target_col in numerical_cols:
        numerical_cols.remove(target_col)

    # Calculate correlations
    corr_0 = target_0[numerical_cols].corr()
    corr_1 = target_1[numerical_cols].corr()

    def get_top_correlations(corr_matrix, n=top_n):
        # Get upper triangle
        upper_tri = corr_matrix.where(
            np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

        # Find top correlations
        correlations = []
        for col in upper_tri.columns:
            for idx in upper_tri.index:
                if pd.notna(upper_tri.loc[idx, col]):
                    correlations.append({
                        'Variable_1': idx,
                        'Variable_2': col,
                        'Correlation': upper_tri.loc[idx, col],
                        'Abs_Correlation': abs(upper_tri.loc[idx, col])
                    })

        corr_df = pd.DataFrame(correlations)
        return corr_df.nlargest(n, 'Abs_Correlation')

    # Get top correlations for each class
    top_corr_0 = get_top_correlations(corr_0, top_n)
    top_corr_1 = get_top_correlations(corr_1, top_n)

    print(f" TOP {top_n} CORRELATIONS - NO PAYMENT DIFFICULTIES:")
    print(top_corr_0[['Variable_1', 'Variable_2', 'Correlation']].to_string(index=False))

    print(f"\n TOP {top_n} CORRELATIONS - PAYMENT DIFFICULTIES:")
    print(top_corr_1[['Variable_1', 'Variable_2', 'Correlation']].to_string(index=False))

    # Visualize correlation heatmaps
    fig, axes = plt.subplots(1, 2, figsize=(20, 8))

    # Select top correlated variables for heatmap
    top_vars = list(set(top_corr_0['Variable_1'].tolist() + top_corr_0['Variable_2'].tolist()))[:15]

    sns.heatmap(corr_0[top_vars].loc[top_vars], annot=True, cmap='coolwarm',
                center=0, ax=axes[0], fmt='.2f')
    axes[0].set_title('Correlation Matrix - No Payment Difficulties')

    sns.heatmap(corr_1[top_vars].loc[top_vars], annot=True, cmap='coolwarm',
                center=0, ax=axes[1], fmt='.2f')
    axes[1].set_title('Correlation Matrix - Payment Difficulties')

    plt.tight_layout()
    plt.show()

    return top_corr_0, top_corr_1


# SECTION 7: SEGMENTED ANALYSIS


def segmented_analysis(df, target_col='TARGET'):
    """Detailed segmented analysis by target variable"""

    print("SEGMENTED ANALYSIS")


    # Key variables for detailed analysis
    key_numerical = ['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY', 'DAYS_BIRTH', 'DAYS_EMPLOYED']
    key_categorical = ['CODE_GENDER', 'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS', 'NAME_INCOME_TYPE']

    # Numerical variables segmented analysis
    print(" NUMERICAL VARIABLES - SEGMENTED ANALYSIS:")

    for col in key_numerical:
        if col in df.columns:
            print(f"\n{col}:")
            stats_by_target = df.groupby(target_col)[col].agg([
                'count', 'mean', 'median', 'std', 'min', 'max'
            ]).round(2)
            print(stats_by_target)

            # Statistical test
            group_0 = df[df[target_col] == 0][col].dropna()
            group_1 = df[df[target_col] == 1][col].dropna()

            if len(group_0) > 0 and len(group_1) > 0:
                statistic, p_value = stats.mannwhitneyu(group_0, group_1)
                print(f"Mann-Whitney U test p-value: {p_value:.6f}")
                if p_value < 0.05:
                    print(" Statistically significant difference between groups")
                else:
                    print(" No statistically significant difference")

    # Categorical variables segmented analysis
    print(f"\n CATEGORICAL VARIABLES - SEGMENTED ANALYSIS:")

    for col in key_categorical:
        if col in df.columns and df[col].nunique() <= 20:
            print(f"\n{col}:")

            # Cross-tabulation
            ct = pd.crosstab(df[col], df[target_col], margins=True)
            ct_pct = pd.crosstab(df[col], df[target_col], normalize='index') * 100

            print("Count Distribution:")
            print(ct)
            print("\nPercentage Distribution:")
            print(ct_pct.round(2))

            # Chi-square test
            chi2, p_value, dof, expected = stats.chi2_contingency(ct.iloc[:-1, :-1])
            print(f"Chi-square test p-value: {p_value:.6f}")
            if p_value < 0.05:
                print(" Statistically significant association with target")
            else:
                print(" No statistically significant association")

    # Visualization of key segments
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))

    # Income distribution by target
    if 'AMT_INCOME_TOTAL' in df.columns:
        axes[0,0].hist(df[df[target_col]==0]['AMT_INCOME_TOTAL'], bins=50, alpha=0.7, label='No Difficulties', density=True)
        axes[0,0].hist(df[df[target_col]==1]['AMT_INCOME_TOTAL'], bins=50, alpha=0.7, label='Payment Difficulties', density=True)
        axes[0,0].set_xlabel('Income')
        axes[0,0].set_ylabel('Density')
        axes[0,0].set_title('Income Distribution by Target')
        axes[0,0].legend()
        axes[0,0].set_xlim(0, df['AMT_INCOME_TOTAL'].quantile(0.95))

    # Credit amount by target
    if 'AMT_CREDIT' in df.columns:
        axes[0,1].hist(df[df[target_col]==0]['AMT_CREDIT'], bins=50, alpha=0.7, label='No Difficulties', density=True)
        axes[0,1].hist(df[df[target_col]==1]['AMT_CREDIT'], bins=50, alpha=0.7, label='Payment Difficulties', density=True)
        axes[0,1].set_xlabel('Credit Amount')
        axes[0,1].set_ylabel('Density')
        axes[0,1].set_title('Credit Amount Distribution by Target')
        axes[0,1].legend()
        axes[0,1].set_xlim(0, df['AMT_CREDIT'].quantile(0.95))

    # Age distribution by target
    if 'DAYS_BIRTH' in df.columns:
        df['AGE'] = abs(df['DAYS_BIRTH']) / 365.25
        axes[1,0].hist(df[df[target_col]==0]['AGE'], bins=30, alpha=0.7, label='No Difficulties', density=True)
        axes[1,0].hist(df[df[target_col]==1]['AGE'], bins=30, alpha=0.7, label='Payment Difficulties', density=True)
        axes[1,0].set_xlabel('Age (Years)')
        axes[1,0].set_ylabel('Density')
        axes[1,0].set_title('Age Distribution by Target')
        axes[1,0].legend()

    # Gender distribution by target
    if 'CODE_GENDER' in df.columns:
        gender_target = pd.crosstab(df['CODE_GENDER'], df[target_col], normalize='index') * 100
        gender_target.plot(kind='bar', ax=axes[1,1])
        axes[1,1].set_title('Payment Difficulties by Gender (%)')
        axes[1,1].set_xlabel('Gender')
        axes[1,1].set_ylabel('Percentage')
        axes[1,1].legend(['No Difficulties', 'Payment Difficulties'])
        axes[1,1].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()

# SECTION 8: BUSINESS INSIGHTS AND RISK FACTORS


def identify_risk_factors(df, target_col='TARGET'):
    """Identify key risk factors for loan default"""

    print("BUSINESS INSIGHTS & RISK FACTORS")


    risk_factors = {}

    # 1. Income to Credit Ratio Analysis
    if 'AMT_INCOME_TOTAL' in df.columns and 'AMT_CREDIT' in df.columns:
        df['INCOME_CREDIT_RATIO'] = df['AMT_INCOME_TOTAL'] / df['AMT_CREDIT']

        ratio_by_target = df.groupby(target_col)['INCOME_CREDIT_RATIO'].agg(['mean', 'median'])
        print(" INCOME TO CREDIT RATIO ANALYSIS:")
        print(ratio_by_target.round(3))

        # Risk threshold
        low_ratio_threshold = df['INCOME_CREDIT_RATIO'].quantile(0.25)
        high_risk_ratio = df[df['INCOME_CREDIT_RATIO'] <= low_ratio_threshold][target_col].mean()
        low_risk_ratio = df[df['INCOME_CREDIT_RATIO'] > low_ratio_threshold][target_col].mean()

        print(f"Default rate for low income/credit ratio ({low_ratio_threshold:.2f}): {high_risk_ratio:.1%}")
        print(f"Default rate for higher income/credit ratio: {low_risk_ratio:.1%}")

        risk_factors['Low Income/Credit Ratio'] = {
            'threshold': low_ratio_threshold,
            'risk_increase': (high_risk_ratio - low_risk_ratio) / low_risk_ratio
        }

    # 2. Age Analysis
    if 'DAYS_BIRTH' in df.columns:
        df['AGE'] = abs(df['DAYS_BIRTH']) / 365.25
        df['AGE_GROUP'] = pd.cut(df['AGE'], bins=[0, 25, 35, 45, 55, 100],
                                labels=['18-25', '26-35', '36-45', '46-55', '55+'])

        age_risk = df.groupby('AGE_GROUP')[target_col].agg(['count', 'mean']).round(3)
        print(f"\n AGE GROUP RISK ANALYSIS:")
        age_risk.columns = ['Count', 'Default_Rate']
        print(age_risk)

        # Identify highest risk age group
        highest_risk_age = age_risk['Default_Rate'].idxmax()
        lowest_risk_age = age_risk['Default_Rate'].idxmin()
        print(f"Highest risk age group: {highest_risk_age} ({age_risk.loc[highest_risk_age, 'Default_Rate']:.1%})")
        print(f"Lowest risk age group: {lowest_risk_age} ({age_risk.loc[lowest_risk_age, 'Default_Rate']:.1%})")

    # 3. Employment Analysis
    if 'DAYS_EMPLOYED' in df.columns:
        # Handle anomalous positive values (should be negative)
        df['DAYS_EMPLOYED_CLEAN'] = df['DAYS_EMPLOYED'].apply(lambda x: x if x < 0 else np.nan)
        df['EMPLOYMENT_YEARS'] = abs(df['DAYS_EMPLOYED_CLEAN']) / 365.25

        employment_stats = df.groupby(target_col)['EMPLOYMENT_YEARS'].agg(['mean', 'median']).round(2)
        print(f"\n EMPLOYMENT EXPERIENCE ANALYSIS:")
        print(employment_stats)

        # Employment stability buckets
        df['EMPLOYMENT_STABILITY'] = pd.cut(df['EMPLOYMENT_YEARS'],
                                           bins=[0, 1, 5, 10, 50],
                                           labels=['<1 year', '1-5 years', '5-10 years', '10+ years'])

        emp_risk = df.groupby('EMPLOYMENT_STABILITY')[target_col].agg(['count', 'mean']).round(3)
        print("Employment Stability Risk:")
        emp_risk.columns = ['Count', 'Default_Rate']
        print(emp_risk)

    # 4. External Source Scores Analysis
    external_cols = [col for col in df.columns if 'EXT_SOURCE' in col]
    if external_cols:
        print(f"\n EXTERNAL SOURCE SCORES ANALYSIS:")

        for col in external_cols:
            if df[col].notna().sum() > 1000:  # Only analyze if sufficient data
                score_stats = df.groupby(target_col)[col].agg(['mean', 'median']).round(3)
                print(f"\n{col}:")
                print(score_stats)

                # Low score risk
                low_score_threshold = df[col].quantile(0.25)
                high_risk_score = df[df[col] <= low_score_threshold][target_col].mean()
                low_risk_score = df[df[col] > low_score_threshold][target_col].mean()

                print(f"Default rate for low {col} ({low_score_threshold:.2f}): {high_risk_score:.1%}")
                print(f"Default rate for higher {col}: {low_risk_score:.1%}")

    # 5. Loan Purpose Analysis
    if 'NAME_CONTRACT_TYPE' in df.columns:
        contract_risk = df.groupby('NAME_CONTRACT_TYPE')[target_col].agg(['count', 'mean']).round(3)
        print(f"\n CONTRACT TYPE RISK ANALYSIS:")
        contract_risk.columns = ['Count', 'Default_Rate']
        print(contract_risk)

    # 6. Education Level Analysis
    if 'NAME_EDUCATION_TYPE' in df.columns:
        education_risk = df.groupby('NAME_EDUCATION_TYPE')[target_col].agg(['count', 'mean']).round(3)
        print(f"\n EDUCATION LEVEL RISK ANALYSIS:")
        education_risk.columns = ['Count', 'Default_Rate']
        print(education_risk)

    return risk_factors


# SECTION 9: PREVIOUS APPLICATION ANALYSIS

def analyze_previous_applications(app_df, prev_df, target_col='TARGET'):
    """Analyze previous application patterns"""
    print("\n" + "="*60)
    print("PREVIOUS APPLICATION ANALYSIS")
    print("="*60)

    # Merge datasets
    prev_summary = prev_df.groupby('SK_ID_CURR').agg({
        'NAME_CONTRACT_STATUS': ['count', lambda x: (x == 'Approved').sum(),
                                lambda x: (x == 'Refused').sum()],
        'AMT_CREDIT': 'mean',
        'AMT_APPLICATION': 'mean'
    }).round(2)

    prev_summary.columns = ['Total_Prev_Apps', 'Approved_Apps', 'Refused_Apps',
                           'Avg_Prev_Credit', 'Avg_Prev_Application']

    # Calculate approval rates
    prev_summary['Approval_Rate'] = (prev_summary['Approved_Apps'] /
                                    prev_summary['Total_Prev_Apps']).fillna(0)
    prev_summary['Refusal_Rate'] = (prev_summary['Refused_Apps'] /
                                   prev_summary['Total_Prev_Apps']).fillna(0)

    # Merge with application data
    app_with_prev = app_df.merge(prev_summary, left_on='SK_ID_CURR',
                                right_index=True, how='left')

    # Fill NaN for customers with no previous applications
    prev_cols = ['Total_Prev_Apps', 'Approved_Apps', 'Refused_Apps',
                'Avg_Prev_Credit', 'Avg_Prev_Application', 'Approval_Rate', 'Refusal_Rate']
    app_with_prev[prev_cols] = app_with_prev[prev_cols].fillna(0)

    print(" PREVIOUS APPLICATION IMPACT ON CURRENT DEFAULT:")

    # Analyze impact of previous applications
    has_prev = app_with_prev['Total_Prev_Apps'] > 0
    no_prev_default_rate = app_with_prev[~has_prev][target_col].mean()
    has_prev_default_rate = app_with_prev[has_prev][target_col].mean()

    print(f"Default rate - No previous applications: {no_prev_default_rate:.1%}")
    print(f"Default rate - Has previous applications: {has_prev_default_rate:.1%}")

    # Approval rate impact
    if has_prev.sum() > 0:
        approval_rate_impact = app_with_prev[has_prev].groupby(
            pd.cut(app_with_prev[has_prev]['Approval_Rate'], bins=[0, 0.5, 0.8, 1.0],
                  labels=['Low (<50%)', 'Medium (50-80%)', 'High (>80%)'])
        )[target_col].agg(['count', 'mean']).round(3)

        print(f"\nDefault rate by previous approval rate:")
        approval_rate_impact.columns = ['Count', 'Default_Rate']
        print(approval_rate_impact)

    return app_with_prev

# SECTION 10: MAIN EXECUTION AND SUMMARY

def generate_business_summary(df, risk_factors, target_col='TARGET'):
    """Generate executive summary for business stakeholders"""

    print("EXECUTIVE SUMMARY & RECOMMENDATIONS")


    total_applications = len(df)
    default_rate = df[target_col].mean()

    print(f" PORTFOLIO OVERVIEW:")
    print(f"Total Applications Analyzed: {total_applications:,}")
    print(f"Overall Default Rate: {default_rate:.1%}")
    print(f"Estimated Financial Impact: ${total_applications * default_rate * 50000:,.0f}")  # Assuming avg loan $50k

    print(f"\n KEY RISK FACTORS IDENTIFIED:")

    # Top risk indicators
    risk_indicators = [
        "Low Income-to-Credit Ratio",
        "Young Age (18-25 years)",
        "Short Employment History (<1 year)",
        "Low External Credit Scores",
        "Previous Loan Rejections"
    ]

    for i, indicator in enumerate(risk_indicators, 1):
        print(f"{i}. {indicator}")

    print(f"\n BUSINESS RECOMMENDATIONS:")

    recommendations = [
        "Implement stricter income verification for high credit amounts",
        "Require additional guarantors for applicants under 25",
        "Consider employment stability as key approval criterion",
        "Integrate external credit bureau scores in decision model",
        "Offer financial literacy programs to high-risk segments",
        "Implement risk-based pricing for borderline applications",
        "Regular model retraining with new data patterns"
    ]

    for i, rec in enumerate(recommendations, 1):
        print(f"{i}. {rec}")



def main_analysis():
    """Main function to run complete EDA analysis"""
    print("Starting Comprehensive Loan Default Risk Analysis")

    # Step 1: Load and explore data
    app_data, prev_data = load_and_explore_data()
    if app_data is None:
        return

    # Step 2: Missing data analysis
    missing_df, high_missing, medium_missing, low_missing = analyze_missing_data(app_data)

    # Step 3: Handle missing data
    app_clean = handle_missing_data(app_data, high_missing)

    # Step 4: Outlier detection
    outlier_df = detect_outliers(app_clean)

    # Step 5: Univariate analysis
    univariate_analysis(app_clean)

    # Step 6: Bivariate analysis
    bivariate_analysis(app_clean)

    # Step 7: Correlation analysis
    top_corr_0, top_corr_1 = correlation_analysis(app_clean)

    # Step 8: Segmented analysis
    segmented_analysis(app_clean)

    # Step 9: Business insights
    risk_factors = identify_risk_factors(app_clean)

    # Step 10: Previous application analysis
    if prev_data is not None:
        app_with_prev = analyze_previous_applications(app_clean, prev_data)
    else:
        app_with_prev = app_clean

    # Step 11: Business summary
    generate_business_summary(app_with_prev, risk_factors)

    print("\n Analysis Complete!")
    return app_with_prev, risk_factors, top_corr_0, top_corr_1

# EXECUTION

if __name__ == "__main__":
    # Run the complete analysis
    final_df, risk_factors, correlations_0, correlations_1 = main_analysis()
```
